# Conversion Between PDF and Image

## JPG to PDF

```python

import img2pdf
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

path = './ocean.pdf'
with open(file=path, mode='wb') as pdf:
    """convert jpg to pdf"""
    pdf.write(img2pdf.convert('./ocean.jpg'))
    pdf.close()
    img = Image.open(fp='./ocean.jpg')
    plt.imshow(X=img)
```

## PNG to PDF
