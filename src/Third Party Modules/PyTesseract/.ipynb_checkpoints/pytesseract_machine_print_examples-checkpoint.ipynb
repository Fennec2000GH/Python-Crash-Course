# OCR with PyTesseract

This notebook deals with only machine-typed text.

```python

import cv2
import matplotlib.pyplot as plt
import pytesseract
%matplotlib inline

path = './einstein_quote.jpg'
img = cv2.imread(filename=path, flags=cv2.IMREAD_GRAYSCALE)

plt.imshow(X=img, aspect='auto', origin='upper', cmap='gray')

result = pytesseract.image_to_string(image=img)
print(result)
```
