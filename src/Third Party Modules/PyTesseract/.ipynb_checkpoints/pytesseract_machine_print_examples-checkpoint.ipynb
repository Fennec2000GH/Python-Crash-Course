# OCR with PyTesseract (Machine-Generated Text)

This notebook deals with only machine-typed text.

```python

import cv2
import matplotlib.pyplot as plt
import pytesseract
from pytesseract import Output
%matplotlib inline

path = './einstein_quote.jpg'
img = cv2.imread(filename=path, flags=cv2.IMREAD_COLOR)
print(img.shape)

# Plotting
fig = plt.figure(num=None, figsize=(20, 20))
plt.imshow(X=img, aspect='auto', origin='upper', cmap='gray')

# Detected words
result = pytesseract.image_to_string(image=img)
print(result)
print('-' * 125)
result_data = pytesseract.image_to_data(image=img, output_type=Output.DICT)
print(result_data)
```
